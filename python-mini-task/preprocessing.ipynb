```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("input.csv", sep = ';', header=None)
data.columns = data.iloc[6,:]+'_['+data.iloc[7,:]+']'
data = data.iloc[9:,].reset_index(drop = True)

data.head(11)
```

```python
# data.to_csv("data2.csv", sep = ";")
data = pd.read_csv("data2.csv", sep=";" , decimal = ",", thousands='.').drop(["Unnamed: 0"], axis=1)
# data.head()
# data.dtypes
```

```python
col_interest = ['VhubC_m/s', 'Omega_rpm', 'MSTor_deg']

fig, axes = plt.subplots(3, 1, figsize=(10, 18))

for i, column in enumerate(col_interest):
    axes[i].plot(data['t_(s)'], data[column], label=column)
    axes[i].set_title(f'Time Series: {column}')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Value')
    
    interval = 100  #Adjust this interval as needed
    
    x_ticks = data['t_(s)'][::100]
    axes[i].set_xticks(x_ticks)
    
    axes[i].legend()

plt.show()
```

```python
# make sure first, that this column represents fractions of a second.
data['timedelta'] = pd.to_timedelta(data["t_(s)"], unit='s')
data.set_index('timedelta', inplace=True)
data.head()
```

```python
data.tail()
```

```python
import seaborn as sns

df = data['MSTor_deg']


# 1. Resampling

plt.figure(figsize=(10, 6))

sampled_data = df.resample('2S').mean()

plt.plot(sampled_data, marker='o')
plt.title('2 Seconds Inteval Samples')
plt.xlabel('Time')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

```python
# 2. Moving Averages
rolling_mean = df.rolling(window=50).mean()

# Plot the rolling mean
plt.figure(figsize=(10, 6))
plt.plot(data['t_(s)'], df, label='Original Data', alpha=0.5)
plt.plot(data['t_(s)'], rolling_mean, label='Rolling Mean', color='orange')


plt.title('Original Data with a 1 second Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()



```
