```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# reading the csv file
data = pd.read_csv("input.csv", sep = ';', header=None)

# making the new header
data.columns = data.iloc[6,:]+'_['+data.iloc[7,:]+']'

# dropping useless rows.
data = data.iloc[9:,].reset_index(drop = True)

data.head()
```

```python
# loading the data, with the right *decimal* & *thousands* arguments.

# un-comment the following line for the first run, then, // 
# once the new csv file is generated, make it a comment again.

#data.to_csv("src/data2.csv", sep = ";")  
data = pd.read_csv("src/data2.csv", sep=";" , decimal = ",", thousands='.').drop(["Unnamed: 0"], axis=1)
data.head()
```

```python
# turning the column t_[(s)]  into a datetime type
data["time_index"] = pd.to_datetime(data["t_[(s)]"], unit='s')

# setting the newely created col. as index. //
# (for conveniency, I decided to keep the col data["t_[(s)"], //
# e.g., for plotting)

data.set_index("time_index", inplace=True)
data.head()
```

```python
# the index is a TimedeltaIndex now. Good.
type(data.index)
```

```python
# saving, the pre-processed data. (set index_label as False)
data.to_csv("src/output.csv", sep = ";", index_label = False)
```

```python
# proper way of leading:(set parse_dates as True, so the index is DatetimeIndex)
data2 = pd.read_csv("src/output.csv", sep = ";", parse_dates=True)
type(data2.index)
```
